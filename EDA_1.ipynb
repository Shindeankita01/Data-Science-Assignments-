### 2. Descriptive Statistics  

```python
round(df.describe(),4)
```

```python
def capping(data,columns):
    for col in columns:
        q1=data[col].quantile(0.25)
        q3=data[col].quantile(0.75)
        iqr=q3-q1
        print(f"*IQR value of {col}:",iqr)
```

```python
df.columns
```

```python
print(capping(df,df.columns))
```

### 3.Data Visualizations:

```python
# Histogram
fig, axes = plt.subplots(3, 4, figsize=(15, 10))
fig.suptitle("Histograms of Numerical Features")
columns = df.columns[:12]
for ax, col in zip(axes.flatten(), columns):
    sns.histplot(df[col], bins=20, kde=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pattern Recognition
correlation_matrix = df.corr()
highly_correlated_pairs = correlation_matrix.unstack().sort_values(ascending=False)
print("Highly Correlated Variable Pairs:")
print(highly_correlated_pairs[highly_correlated_pairs < 1].head(10))
```

#### Correlation Insights:
* Strong Positive Correlations:
  
  Width & MSTV (r = 0.44) → Suggests that the overall variability of fetal heart rate is related to     short-term variability.

  NSP & ASTV (r = 0.42) → Indicates that abnormal short-term variability is associated with non-stress   test results.

  DL & MSTV (r = 0.41) → Suggests that decelerations in fetal heart rate are linked with short-term     variability.

* Moderate Positive Correlations:

  NSP & ALTV (r = 0.37) → Abnormal long-term variability is related to non-stress test results.

  Width & DL (r = 0.41) → Decelerations affect the overall width of heart rate variability.

### Key Insights from Exploratory Data Analysis (EDA):

  - AC(Accelerations), FM(Fetal Movements), DL(Late Decelerations), and DS(Short Decelerations) have distributions                   concentrated around zero with a long tail.
  - This suggests these fetal distress indicators are rare in the dataset.
  - ASTV(Abnormal Short-Term Variability), UC(Uterine Contractions), and Width exhibit multimodal distributions, indicating         the presence of different fetal health conditions (e.g., normal vs. distressed).
  - Some variables, like DL, UC, and Width, have negative values, which might be due to data entry errors or incorrect               measurements.
  - Extremely high values in ASTV, MLTV, and Width might indicate high-risk pregnancies and warrant further investigation.
  - DS (Short Decelerations) and DP (Prolonged Decelerations) have almost all values concentrated at zero.
